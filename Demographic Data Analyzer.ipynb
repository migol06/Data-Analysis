```python
import pandas as pd
```

```python
df = pd.read_csv('adult.data.csv')
```

```python
race_count = df['race'].value_counts()
```

```python
print("Number of each race:\n", race_count) 
```

```python
average_age_men = round(df[df['sex']=='Male']['age'].mean(),1)
```

```python
print("Average age of men:", average_age_men)
```

```python
percentage_bachelors = round(df[df['education']=='Bachelors'].shape[0]/df.shape[0]*100,1)
```

```python
print(f"Percentage with Bachelors degrees: {percentage_bachelors}%")
```

```python
higher_education = df[df['education'].isin(['Bachelors','Masters', 'Doctorate'])]
```

```python
lower_education = df[~df['education'].isin(['Bachelors','Masters', 'Doctorate'])]
```

```python
higher_education_rich = round(higher_education[higher_education['salary'] == '>50K']['salary'].count()/higher_education.shape[0]*100,1)
```

```python
(f"Percentage with higher education that earn >50K: {higher_education_rich}%")
```

```python
 lower_education_rich = round(lower_education[lower_education['salary']=='>50K']['salary'].count()/lower_education.shape[0]*100,1)
```

```python
(f"Percentage without higher education that earn >50K: {lower_education_rich}%")
```

```python
min_work_hours = df['hours-per-week'].min()
```

```python
print(f"Min work time: {min_work_hours} hours/week")
```

```python
num_min_workers = df[df['hours-per-week']==1]['hours-per-week'].count()
```

```python
rich_percentage = df[(df['hours-per-week']==1) & (df['salary']=='>50K')].shape[0]/num_min_workers*100
```

```python
print(f"Percentage of rich among those who work fewest hours: {rich_percentage}%")
```

```python
highest_earning_country = round((df[df['salary']=='>50K']['native-country'].value_counts() / df['native-country'].value_counts())*100,1).idxmax()
```

```python
print("Country with highest percentage of rich:", highest_earning_country)
```

```python
highest_earning_country_percentage = round((df[df['salary']=='>50K']['native-country'].value_counts()/df['native-country'].value_counts())*100,1).max()
```

```python
print(f"Highest percentage of rich people in country: {highest_earning_country_percentage}%")
```

```python
top_IN_occupation = df[(df['salary']==">50K") & (df['native-country']=='India')].groupby('occupation')['occupation'].count().idxmax()
```

```python
print("Top occupations in India:", top_IN_occupation)
```
